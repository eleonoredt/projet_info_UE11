Simulation de systèmes dynamiques non lisses
==============================


Le cours se focalise sur l'étude de systèmes dynamiques réguliers, c'est-à-dire $\dot{x}=f(x)$ avec $f$ au moins continue, et le plus souvent continûment différentiable. Le but de ce projet est de découvrir des systèmes moins réguliers, dits *non-lisses*. Il existe une myriade de modélisations de ces systèmes mais on peut distinguer différents *niveaux* d'irrégularité:

 - $f$ est continue mais non différentiable en certains points (on parle de *nonsmooth dynamics of type 1*, i.e., *NSD1*) : cela peut arriver dans la modélisation d'écoulements par exemple. Dans ce cas, les solutions $t\mapsto x(t)$ restent $C^1$ mais leur simulation peut déjà se compliquer.

 - $f$ est discontinue en certains points, ou le long de certaines surfaces (on parle de *NSD2*) : ceci apparaît typiquement en mécanique lors de la modélisation de contacts et friction entre solides, ou dans la modélisation de circuits électriques contenant des diodes par exemple. Dans ce cas, les solutions $t\mapsto x(t)$ sont toujours (absolument) continues mais non différentiables, le schéma numérique doit détecter des changements de *mode*.

 - les solutions $t\mapsto x(t)$ sont elles-mêmes discontinues, présentant des *sauts*  à certains instants ou lorsqu'elles rencontrent certaines surfaces, et il faut ajouter à $f$ un modèle de ces sauts (on parle de *NSD3*) : ceci apparaît typiquement avec les systèmes mécaniques à impact.

L'étude mathématique de ces sytèmes est très complexe, faisant appel à des outils qui nous mèneraient trop loin de ce cours (inclusions différentielles, complémentarité, analyse convexe, *measure differential inclusions*, etc), notre but est seulement d'en découvrir certains aspects à travers des exemples iconiques. ## Dynamique non différentiable (NSD1)

Reprenons la loi de Toriccelli étudiée en TD et modélisant (modulo reparamétrisation) l'écoulement d'un fluide dans un réservoir
$$
\dot{x} = - \sqrt{|x|} \tag{1}
$$

**Question 1** Coder un schéma Euler explicite permettant de simuler une équation différentielle $\dot{x}=f(x,t)$ à partir d'une condition initiale $x_0$ au temps $t_0$ jusqu'au temps $t_f$ avec un pas de temps $dt$. 

```python
def euler_explicit(f,x0,t0,tf,dt):
    sol=[x0]
    time= np.linspace(t0, tf, int((tf - t0) / dt) + 1)
    for i in range(len(time)-1):
        sol.append(sol[-1]+dt*f(sol[-1],time[i]))
    return time, np.array(sol)
```

**Question 2** L'utiliser pour simuler un écoulement d'une hauteur initiale $x_0=10$ pendant 10 unités de temps. Tracer la solution obtenue. Est-elle celle que l'on attend ? Les solver ''odeint'' ou ''solve_ivp" de scipy fait-il mieux ? Expliquer pourquoi. Proposer une façon de simuler *la* solution *physique*.



```python
def f1(x,t):
    return -np.sqrt(np.abs(x))
    
def f1_ivp(t, x):
    return f1(x,t)

# On est obligées de définir deux fois la fonction f car odeint et euler_explicit prennent f(x,t) en argument tandis que solve_ivp prend f(t,x)

x0=10
t0=0
tf=10
dt=0.1

#Euler explicite
t1, x1 = euler_explicit(f1, x0, t0, tf, dt)
plt.plot(t1, x1, label="Euler explicite")

#odeint
t2 = np.arange(t0, tf, dt)
x2 = odeint(f1, x0, t2)
plt.plot(t2,x2,label='odeint', ls='--')

#solve_ivp
sol = solve_ivp(f1_ivp, [t0,tf], [x0])
plt.plot(sol.t,sol.y[0],label='solve_ivp', ls=':')

plt.title(r"$\dot{x} = - \sqrt{|x|}$")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.grid()
plt.show()
```

>La solution qu'on obtient n'est pas physique, car $x(t)$ devient négative au lieu de décroitre puis de rester à 0 au bout d'un certain temps.
Comme vu en TD, cela est lié au fait que la fonction n'est pas différentiable en 0 (ni même lipschitzienne) donc l'unicité de la solution maximale ne peut être garantie par le théorème de Cauchy-Lipschitz. Ainsi les 3 méthodes nous ont donné une des solutions possibles, mais pas celle qui est physique.
>
>Pour simuler la solution physique (où $x(t)=0$ après que le réservoir se soit vidé), on pourrait modifier la fonction pour qu'elle s'annule si $x$ devient négatif en mettant par exemple $
\dot{x} = - \sqrt{\text{max}(0,x)}
$




En pratique, on essaye toujours de se ramener à une modélisation *bien posée* dont les solutions existent et sont uniques (ajout de contraintes ou régularisation, par exemple). Ici, on pourrait par exemple considérer plutôt
$$
\dot{x} = - {\rm sign}(x) \sqrt{|x|} \tag{2}
$$
avec ${\rm sign}(0)$ arbitraire. 

**Question 3** Montrer que pour toute condition initiale dans $\mathbb{R}$, (2) admet une unique solution maximale en temps positif définie sur $\mathbb{R}_{\geq 0}$, et qu'elle correspond à la solution ``physique'' attendue précédemment pour des conditions initiales positives. Justifier que l'on parle de "convergence globale vers 0 en temps fini". Illustrer ce phénomène en simulation. Ce comportement serait-il possible pour $\dot{x}=f(x)$ avec $f$ continûment différentiable ?


**<ins>Réponse 3:</ins>**

On a le problème suivant:


$$
\begin{cases}
\dot{x}= - {\rm sign}(x) \sqrt{|x|} \\
 x(0)=x_0 \in \mathbb{R}
\end{cases}
$$

**Résolution:**
- Supposons que $x_0>0$, en séparant les variables on a:

$$\frac{dx}{\sqrt{x}}= -dt \\
2(\sqrt{x}-\sqrt{x_0})= -t \\
x(t)= (\sqrt{x_0}-\frac{t}{2})^2 $$

Ce raisonnement est valable tant que $\sqrt{x_0}-\frac{t}{2}\geq0 $ i.e. $t \in [0,2\sqrt{x_0}]$


- En effectuant le même raisonnement pour $x_0<0$, on obtient $x(t)= -(\sqrt{|x_0|}-\frac{t}{2})^2 $ pour $t \in [0,2\sqrt{|x_0|}]$

- On peut se dire qu'on prolonge $x$ par $x(t)=0$ pour tout $ t>T=2\sqrt{|x_0|}$.


**Unicité:**
Montrons que cette fonction est unique. 

- Le théorème de Cauchy-Lipschitz nous assure l'existence et l'unicité tant que $x(t) \neq 0$. Une fois que la solution atteint 0, à l'instant $T$, nous ne pouvons utiliser ce théorème car les hypothèses d'application ne sont plus vérifiées (fonction non localement lipschitzienne).

- Montrons que l'unique prolongation pour $t \geq T$ est la fonction nulle ($x(t)=0$). 

Supposons, par l'absurde, qu'il existe une autre prolongation $y$ qui ne reste pas nulle. Ainsi, il existe $t^*>T$ tel que $y(t^*) \neq 0$.

Supposons que $y(t^*)>0$. Alors par continuité, il existe un intervalle après T où $y$ est strictement positive. De plus, $\dot{y}(t)= - \sqrt{y} <0$ sur cet intervalle i.e. $y$ est strictement décroissante. Une fonction qui part de 0 et qui a une dérivée négative ne peut pas devenir strictement positive: il y a donc contradition.

On aurait la même contradiction si on avait supposé que $y(t^*)<0$ après $T$, donc aucune trajectoire ne peut quitter 0 une fois qu'elle l'atteint une première fois. Ainsi, la fonction nulle est l'unique solution possible pour $t \geq T$.

**Conclusion:**
Ainsi, nous avons trouvé l'unique solution maximale en temps positif définie sur $\mathbb{R}_{\geq 0}$. On remarque cette fois-ci qu'on a bien une solution "physique", dans le sens où notre solution tend vers 0 en temps fini, comme le ferait un liquide dans un réservoir. On a donc effectivement une convergence globale en temps fini.

**Et si $f$ était continûment différentiable ?**
Ce comportement aurait été impossible. Pour un tel système régulier, l'unicité des solutions implique qu'une trajectoire ne peut jamais atteindre un point d'équilibre $x^*$en temps fini (en effet, si elle l'atteignait, elle coïnciderait en un point à la solution constante $y \equiv x^*$, or le théorème de Cauchy-Lipschitz assure l'unicité de la solution maximale). C'est la singularité de la racine carrée en 0 qui permet ce phénomène "physique". 


```python
def f2(x,t):
    return -np.sign(x)*np.sqrt(np.abs(x))
    
def f2_ivp(t, x):
    return f2(x,t)
    
x0=10
t0=0
tf=10
dt=0.1

#Euler explicite
t1, x1 = euler_explicit(f2, x0, t0, tf, dt)
plt.plot(t1, x1, label="Euler explicite")

#odeint
t2 = np.arange(t0, tf, dt)
x2 = odeint(f2, x0, t2,mxstep=5000)
#on impose un nombre d'itérations maximal sinon on a une alerte "Excess work done" (ce qui illustre la difficulté numérique de notre problème, le solveur adapte son pas de temps à l'extrême jusqu'à dépasser sa limite de calcul) et le résultat est changeant
plt.plot(t2,x2,label='odeint', ls='--')

#solve_ivp
sol = solve_ivp(f2_ivp, [t0,tf], [x0])
plt.plot(sol.t,sol.y[0],label='solve_ivp',ls=':')

plt.title(r"$\dot{x} = - sign(x)\sqrt{|x|}$")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.grid()
plt.show()
```

## Dynamique discontinue (NSD2)



Passons maintenant à une irrégularité plus forte, lorsque la fonction $f$ décrivant l'équation différentielle $\dot{x}=f(x)$ est discontinue, c'est-à-dire lorsque la solution $x$ suit des dynamiques différentes selon les régions de l'espace (par exemple selon le sens du courant pour des diodes, selon le sens de la vitesse pour des frottements secs etc.). Un exemple emblématique de ce cas de figure est 
$$
\dot{x} = - k {\rm sign}(x) \tag{3}
$$
pour $k>0$ et la valeur de ${\rm sign}(0)$ à définir.

**Question 4** Démontrer que pour une valeur de ${\rm sign}(0)$ bien choisie, (3) admet, pour toute condition initiale $x_0\in \mathbb{R}$, une unique solution maximale $t\mapsto x(t)$ en temps positif définie sur $\mathbb{R}_{\geq 0}$ et convergeant en temps fini vers 0. L'expliciter.

**<ins>Réponse 4:</ins>**

On considère l’équation différentielle

$$
\dot{x}(t) = -k\,\mathrm{sign}(x), \quad k > 0.
$$

On choisit la fonction signe définie par

$$
\mathrm{sign}(x) =
\begin{cases}
1 & \text{si } x > 0, \\
0 & \text{si } x = 0, \\
-1 & \text{si } x < 0.
\end{cases}
$$

Le choix de $\mathrm{sign}(0) = 0$, garantit que $x(t)=0$ est une solution stationnaire.

Soit $x_0 \in \mathbb{R}$ une condition initiale.

- Si $x_0 > 0$, tant que $x(t) > 0$ on a $\mathrm{sign}(x(t)) = 1$ et l’équation devient

$$
\dot{x}(t) = -k.
$$

En intégrant, on obtient

$$
x(t) = x_0 - k t.
$$

Cette solution s’annule au temps

$$
T = \frac{x_0}{k}.
$$

Pour tout $t \ge T$, on a $x(t)=0$ et donc

$$
\dot{x}(t) = -k\,\mathrm{sign}(0) = 0,
$$

ce qui implique que la solution reste identiquement nulle.

- Si $x_0 < 0$, tant que $x(t) < 0$ on a $\mathrm{sign}(x(t)) = -1$ et l’équation devient

$$
\dot{x}(t) = k.
$$

On obtient alors

$$
x(t) = x_0 + k t,
$$

qui s’annule au temps

$$
T = \frac{|x_0|}{k}.
$$

Pour tout $t \ge T$, la solution reste égale à zéro.

- Enfin, si $x_0 = 0$, on a directement

$$
\dot{x}(t) = -k\,\mathrm{sign}(0) = 0,
$$

et donc

$$
x(t) = 0 \quad \text{pour tout } t \ge 0.
$$


**Conclusion:** 
$$
x(t) = \begin{cases} 
x_0 - k \, \text{sign}(x_0) \, t & \text{si } 0 \le t \le \frac{|x_0|}{k} \\ 
0 & \text{si } t \ge \frac{|x_0|}{k} 
\end{cases}
$$

Cette solution est définie pour tout $t \ge 0$, elle est unique (il suffit de procéder de la même manière que pour la question 3: Cauchy-Lipschitz pour $t < \frac{|x_0|}{k}$ puis raisonner par l'absurde et trouver une contradiction entre le signe de la fonction et celui de sa dérivée), et elle converge vers $0$ en temps fini, avec un temps de convergence égal à

$$
T = \frac{|x_0|}{k}.
$$

Considérons maintenant $\delta_m>0$ et l'équation différentielle perturbée
$$
\dot{x} = - k{\rm sign}(x) + \delta(t) \tag{4}
$$
pour $t\mapsto \delta(t)$ telle que $|\delta(t)|\leq \delta_m$ pour tout $t$.

**Question 5** Démontrer que si $k>\delta_m$, alors (4) admet toujours, pour toute condition initiale $x_0\in \mathbb{R}^*$ et quelle que soit la perturbation $\delta$ bornée par $\delta_m$, une unique solution maximale $t\mapsto x(t)$ en temps positif et convergeant en temps fini vers 0. Justifier que ces solutions pourraient être définies sur $\mathbb{R}_{\geq 0}$ si on acceptait que ${\rm sign}(0)$ puisse prendre n'importe quelle valeur entre -1 et 1.

**<ins>Réponse 5:</ins>**

On considère l’équation différentielle perturbée et on suppose dans toute la suite que $k > \delta_m$ et que la condition initiale $x_0 \in \mathbb{R}^*$.

- Soit tout d’abord le cas $x_0 > 0$:

Tant que $x(t) > 0$, on a $\mathrm{sign}(x(t)) = 1$ et l’équation devient

$$
\dot{x}(t) = -k + \delta(t).
$$

Par l’hypothèse $|\delta(t)| \le \delta_m$, on obtient l’inégalité

$$
\dot{x}(t) \le -k + \delta_m < 0.
$$

La fonction $x(t)$ est donc strictement décroissante tant qu’elle est positive. En intégrant l’inégalité précédente, on obtient

$$
x(t) \le x_0 - (k - \delta_m)t.
$$

Il existe donc un temps fini

$$
T \le \frac{x_0}{k - \delta_m}
$$

tel que $x(T)=0$. La solution atteint 0 en temps fini. 

- De même, si $x_0 < 0$, tant que $x(t) < 0$ on a $\mathrm{sign}(x(t)) = -1$ et l’équation devient

$$
\dot{x}(t) = k + \delta(t).
$$

On en déduit

$$
\dot{x}(t) \ge k - \delta_m > 0,
$$

ce qui implique que $x(t)$ est strictement croissante tant qu’elle est négative. On obtient alors

$$
x(t) \ge x_0 + (k - \delta_m)t,
$$

et la solution atteint également $0$ en un temps fini

$$
T \le \frac{|x_0|}{k - \delta_m}.
$$

Ainsi, quelle que soit la perturbation $\delta$ vérifiant $|\delta(t)| \le \delta_m$, la solution atteint $0$ en temps fini pour toute condition initiale $x_0 \in \mathbb{R}^*$. On peut prolonger la solution par la fonction nulle pour tout $t > T$ et montrer l'unicité de cette solution maximale de la même manière que pour les questions 3 et 4 (thèorème de Cauchy-Lipschitz pour $t<T$ puis raisonnement par l'absurde).



Si l’on accepte que la fonction signe prenne à l’origine n’importe quelle valeur dans l’intervalle $[-1,1]$, on est conduit à considérer l’inclusion différentielle

$$
\dot{x}(t) =-k\,\mathrm{Sign}(x(t)) + \delta(t),
$$

où la fonction signe multivaluée est définie par

$$
{\rm Sign}(x) = \left\{
\begin{array}{ll}
{1} & \text{si x>0} \\
[-1,1] & \text{si x=0} \\
-1 & \text{si x<0}
\end{array}
\right.
$$

Dans ce cadre, il est toujours possible de choisir une valeur de $\mathrm{Sign}(0)$ compensant exactement la perturbation $\delta(t)$, ce qui permet de maintenir $\dot{x}(t)=0$ lorsque $x(t)=0$. 

En effet, on veut $k\mathrm{Sign}(0) = \delta(t)$, si on prend $\mathrm{Sign}(0)=\frac{\delta(t)}{k}$, comme $k > \delta_m \geq |\delta(t)|$, on a bien $\mathrm{Sign}(0) \in [-1,1]$

Les solutions peuvent alors être prolongées pour tout $t \ge 0$ et rester à l’origine après l’avoir atteinte en temps fini.


Ceci mène en fait à la notion d'*inclusion différentielle* $\dot{x}\in f(x)$, où la fonction signe serait ici une fonction *multivaluée*, c'est-à-dire à valeurs ensembliste (*set-valued map*), définie par
$$
{\rm Sign}(x) = \left\{
\begin{array}{ll}
{1} & \text{si x>0} \\
[-1,1] & \text{si x=0} \\
-1 & \text{si x<0}
\end{array}
\right.
$$
Ce genre de phénomène est utilisé en commande ou estimation par *mode-glissant* (*sliding mode*) pour stabiliser des systèmes dynamiques en temps fini et malgré la présence de perturbations inconnues ou d'incertitudes dans le modèle. Le prix à payer pour ce genre de système est une grande sensibilité au bruit et une difficulté accrue de les discrétiser et simuler.

*Pour la culture* : La théorie des inclusions différentielles a été développée par Filippov, voir https://link.springer.com/book/10.1007/978-3-319-28664-8. Elle a montré que le ``bon'' cadre pour qu'une inclusion différentielle ait de bonnes propriétés (notamment que des approximations de solutions convergent vers des solutions, c'est-à-dire la compacité séquentielle des solutions, permettant la discrétisation et simulation de ces systèmes) est lorsque $f$ est semi-continue supérieure, localement bornée et à valeurs convexes et fermées. D'où l'intervalle $[-1,1]$ pour ${\rm sign}(0)$ qui est le plus petit convexe fermé assurant la semi-continuité supérieure de ${\rm sign}$. 


**Question 6** Illustrer numériquement la convergence en temps finie globale et la robustesse aux perturbations de (4) à l'aide du schéma numérique de votre choix.

```python
k = 1.0
delta_m = 0.3
dt = 1e-3
T = 5.0

def delta(t):
    return delta_m * np.sin(5*t) #avec ce choix, on respecte bien la condition  |delta(t)| inférieure ou égale à delta_m pour tout t

def f3(x,t):
    return -k * np.sign(x) + delta(t) 
   
x0_list = [-3, 0, 3]

plt.figure(figsize=(8,5))
for x0 in x0_list:
    t, x = euler_explicit(f3, x0, 0, T, dt)
    plt.plot(t, x, label=f"$x_0={x0}$")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title(r"$\dot{x}(t) = - ksign(x(t))+\delta(t)$ où $k=1$ et $\delta(t)=0.3sin(5t)$")
plt.legend()
plt.grid()
plt.show()
```

**Question 7** Illustrer et commenter le comportement du schéma d'Euler explicite autour de 0. On parle de phénomène de chattering. Que se passe-t-il si $k$ est grand ? Proposer un schema d'Euler implicite pour ce système et montrer qu'il permet d'éviter ce *chattering* et d'assurer une convergence numérique exacte vers 0.


**<ins>Réponse 7</ins>**

Lorsque l’on applique le schéma d’Euler explicite, on obtient la récurrence

$$
x_{n+1} = x_n +\Delta t(-k\mathrm{sign}(x_n) + \delta(t_n))
$$


Lorsque la solution $x_n$ s'approche de 0, si le pas de temps est trop grand, le terme $-k \Delta t$ devient plus grand que $x_n$ et le schéma devient donc instable (car la distance restante à parcourir est plus petite que la distance minimale que le schéma numérique est obligé de parcourir en un pas). Ainsi, au lieu de converger, la solution traverse brutalement l'axe des abscisses à chaque itération et le système oscille donc indéfiniment autour de 0. 



```python
k_list = [0.6,2,3]  
delta_m = 0.3         
x0 = 1.5   
t0=0
tf=5
dt = 0.1

plt.figure(figsize=(10, 5))
for k in k_list:
    t, x = euler_explicit(f3, x0, t0, tf, dt)
    plt.plot(t, x, label=f'Euler explicite (k={k}, dt={dt})', marker='o', ls='--')
plt.title("Phénomène de chattering avec Euler explicite")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid()
plt.legend()
plt.show()
```

On remarque bien:
- **Mécanisme :** Si $x_n$ est une petite valeur positive, le terme $-k\Delta t$ est plus grand en valeur absolue que $x_n$, propulsant $x_{n+1}$ dans les négatifs. À l'étape suivante, le signe change et le renvoie dans les positifs.
- **Influence de $k$ :** Plus $k$ est grand, plus la "force" de rappel est violente. Pour un pas de temps $\Delta t$ fixe, une augmentation de $k$ augmente l'amplitude des oscillations.

On peut tenter de résoudre un schéma d'Euler implicite pour éviter le problème de chattering: 

$$x_{n+1} = x_n + \Delta t(-k\text{sign}(x_{n+1})+ \delta(t_{n+1}))$$

Pour résoudre cette équation, on utilise la fonction signe multivaluée définie à la question 5 : 
$\text{sign}(0) \in [-1, 1]$.

On pose $\tilde x = x_n + \Delta t\cdot \delta(t_{n+1})$. L'équation peut être réécrite comme :
$$x_{n+1}- \tilde x \in [-k \Delta t, k \Delta t]$$

Deux cas se présentent alors :
1.  **$|\tilde x| \leq k\Delta t$ :** En prenant $x_{n+1} = 0$, l'équation devient $0 + \Delta t \cdot k \cdot e = \tilde x$ avec $e \in [-1, 1]$. Comme $|\tilde x| \le k\Delta t$, il existe bien $e$ qui vérifie bien l'égalité.
2.  **$|\tilde x| > k\Delta t$ :** $k\Delta t$ n'est pas suffisant pour compenser $\tilde x$, la solution $x_{n+1}$ sera donc non nulle et vaudra $x_{n+1}=\tilde x- \text{sign}(\tilde x ) \cdot k\Delta t$



```python
def euler_implicit_sliding(x0, t0, tf, dt):
    time = np.linspace(t0, tf, int((tf - t0) / dt) + 1)
    sol = [x0]
    for i in range(len(time) - 1):
        x_tilde = sol[i] + dt * delta(time[i])
        seuil = k * dt
        if abs(x_tilde) <= seuil:
            sol.append(0.0)
        else:
            sol.append(x_tilde - seuil * np.sign(x_tilde))       
    return time, sol
```

```python
x0 = 3   
t0=0
tf=5
dt = 0.1
k =20.0   # on choisit intentionnellement k grand

t_expl, x_expl = euler_explicit(f3, x0, t0, tf, dt)
t_impl, x_impl = euler_implicit_sliding(x0, t0, tf, dt)


plt.figure(figsize=(10, 6))
plt.plot(t_expl, x_expl, label='Euler explicite', marker='o', ls='--',lw=2)
plt.plot(t_impl, x_impl,  label='Euler implicite', marker='o', ls='--')

plt.title(rf"Comparaison Explicite vs Implicite ($k={k}, dt={dt}$)")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.grid()
plt.show()
```

>Le schéma d’Euler implicite supprime complètement le phénomène de chattering.
Dès que la trajectoire numérique entre dans la bande $|x| \le k\Delta t$,
elle est projetée exactement sur $0$ et y reste.
>
>Ce schéma respecte donc la convergence en temps fini du système continu
et assure une convergence numérique exacte vers $0$, indépendamment de la valeur de $k$.


**Question 8** Application : donner un exemple physique d'un oscillateur évoluant selon
$$
\begin{array}{rcl}
\dot{x}_1 &=& x_2 \\
\dot{x}_2 &=& -x_1 - k {\rm sign}(x_2)  
\end{array}
$$
et proposer un schéma numérique pour simuler son comportement.

**<ins>Réponse 8</ins>**

Le système proposé peut par exemple modéliser un une masse  attachée à un ressort et soumise  frottement sec où:

- **$x_1$** est la position d'une masse $m$ attachée à un ressort
- **$x_2$** est la vitesse de la masse $m$ 
- **$-x_1$** est la force de rappel du ressort.
- **$-k \text{sign}(x_2)$** est la force de frottement solide constante en norme et opposée au mouvement (signe de la vitesse $x_2$)


L'oscillateur va perdre de l'énergie linéairement (et non exponentiellement comme avec des frottements fluides). Le mouvement s’arrête alors en temps fini :la vitesse devient nulle et la masse reste bloquée.

Pour simuler correctement ce comportement et éviter les oscillations numériques (chattering) autour de
$x_2 = 0$ (à faible vitesse), on utilise un schéma d’Euler implicite.

Le schéma s’écrit :

1. **Vitesse :** $x_2^{n+1} = x_2^n + \Delta t \cdot (-x_1^n - k \text{sign}(x_2^n))$ 
2. **Position :** $x_1^{n+1} = x_1^n + \Delta t \cdot x_2^{n+1}$

Le terme discontinu est traité par un raisonnement par cas, comme dans la question précédente, ce qui permet d'éviter le chattering et d’assurer un arrêt exact de la vitesse.

```python
x0 = [10.0,0.0]
t0 = 0.0
tf = 20.0
dt = 0.01
k = 1.0

def euler_implicit_friction(x0, t0, tf, dt):
    time = np.linspace(t0, tf, int((tf - t0) / dt) + 1)
    sol = np.zeros((len(time), 2))
    sol[0] = x0
    for i in range(len(time) - 1):
        curr_x1 = sol[i, 0] 
        curr_x2 = sol[i, 1] 
        v_tilde = curr_x2 + dt * (-curr_x1)
        seuil = k * dt
        if abs(v_tilde) <= seuil:
            next_x2 = 0.0
        else:
            next_x2 = v_tilde - seuil * np.sign(v_tilde)
        next_x1 = curr_x1 + dt * next_x2
        sol[i+1] = [next_x1, next_x2]

    return time, sol


time, sol = euler_implicit_friction(x0, t0, tf, dt)
x1 = sol[:, 0]
x2 = sol[:, 1]

plt.figure(figsize=(10, 5))
plt.plot(time, x1, label="Position $x_1$")
plt.plot(time, x2, label="Vitesse $x_2$")
plt.title(f"Oscillateur avec frottement solide ($k={k}$)")
plt.axhline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```

La simulation montre des oscillations amorties suivies d’un arrêt en temps fini. La vitesse atteint exactement zéro et reste nulle, ce qui correspond au phénomène d’adhérence dû aux frottements solides.

Le schéma implicite permet ainsi de reproduire fidèlement le comportement physique de l’oscillateur et d’éviter les oscillations numériques non physiques autour de $x_2 = 0$.
